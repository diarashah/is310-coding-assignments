# National Parks EDA

This notebook explores U.S. National Parks data using Pandas. 
The goal is to answer key questions about visits and park distribution.

```python
import pandas as pd

parks_data_df = pd.read_csv(
    'https://raw.githubusercontent.com/melaniewalsh/responsible-datasets-in-context/main/datasets/national-parks/US-National-Parks_RecreationVisits_1979-2023.csv'
)
```

```python
parks_data_df.head()
parks_data_df.info()
parks_data_df.describe()
```

```python
import re

def split_on_uppercase(name):
    return re.sub(r'(?<!^)(?=[A-Z])', '_', name).lower()

parks_data_df.columns = [split_on_uppercase(col) for col in parks_data_df.columns]
```

```python
parks_data_df.groupby('state')['recreation_visits'].mean().reset_index()
```

```python
parks_data_df.groupby('park_name')['recreation_visits'].mean().reset_index()
```

```python
parks_data_df.groupby('state')['park_name'].nunique().reset_index()
```
